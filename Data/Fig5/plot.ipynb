```python
import pandas as pd
```

```python
df_CCR5 = pd.read_csv('4BC_CCR5_TM_parsed.txt', names=['model_id', 'tm_active', 'tm_inactive'], skiprows=1, sep=' ')
df_CCR5.info()
df_CCR5.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 散布図の作成
sns.scatterplot(
    x="tm_inactive",
    y="tm_active",
    data=df_CCR5,
    palette='viridis',
    alpha=0.8,
    s=10,

)

# 横線と縦線の追加
# plt.axhline(y=0.86337, linestyle="dotted", color="black", linewidth=0.25)
# plt.axvline(x=0.86337, linestyle="dotted", color="black", linewidth=0.25)

# 軸の範囲の設定
plt.xlim(0.70, 1)
plt.ylim(0.70, 1)

# 軸ラベルの設定
plt.xlabel("Similarity to inactive conformer (TM-score)", fontsize=8, fontweight="bold")
plt.ylabel("Similarity to active conformer (TM-score)", fontsize=8, fontweight="bold")

# グラフのスタイルの設定
sns.set_style("ticks")
sns.despine()
plt.show()
```
